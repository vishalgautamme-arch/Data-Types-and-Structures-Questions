

# THEORY QUESTIONS

### **1. What are data structures, and why are they important?**

Data structures are different ways of storing and organizing data.
They are important because they make our programs faster and easier to work with.



### **2. Difference between mutable and immutable types with examples**

* **Mutable:** Can be changed. Example: list → `[1, 2, 3]`
* **Immutable:** Cannot be changed. Example: string → `"hello"`



### **3. Difference between lists and tuples**

Lists can change (mutable), but tuples cannot change (immutable).
Also, lists use `[]` and tuples use `()`.



### **4. How dictionaries store data**

Dictionaries store data in **key-value pairs** like `"name": "Vishal"`.
Each key points to a value.



### **5. Why use a set instead of a list?**

Use a set when you don’t want duplicates and order doesn’t matter.



### **6. What is a string in Python and how is it different from a list?**

A string is text (like `"hello"`).
You cannot modify a string, but you can modify a list.



### **7. How do tuples ensure data integrity?**

Tuples cannot be changed, so data stays safe and unmodified.



### **8. What is a hash table, and how does it relate to dictionaries?**

Hash table stores data using a hash function.
Python dictionaries use hash tables internally.



### **9. Can lists contain different data types?**

Yes, lists can contain anything: `[1, "hello", 3.5]`.



### **10. Why strings are immutable?**

Because Python stores them in memory in a fixed way for speed and safety.



### **11. Advantages of dictionaries over lists**

Dictionaries make searching fast using keys, unlike lists where you search by index.



### **12. When is tuple better than list?**

When you want data that should not change, like coordinates `(5, 10)`.



### **13. How do sets handle duplicate values?**

They simply remove duplicates automatically.



### **14. How does “in” keyword work differently for lists and dictionaries?**

* In list: checks if item exists
* In dictionary: checks only keys, not values


### **15. Can you modify tuple elements? Why?**

No. Tuples are immutable.



### **16. What is a nested dictionary? Example.**

A dictionary inside another dictionary.
Example:
`{"person": {"name": "Vishal", "age": 20}}`



### **17. Time complexity of accessing dictionary elements**

O(1) → very fast constant time.



### **18. When are lists preferred over dictionaries?**

When order matters or when you only need simple sequences.



### **19. Why are dictionaries unordered?**

They store data by hashing, not by position.
So order is not guaranteed (in concept).



### **20. Difference between list & dictionary in data retrieval**

* List → access by index
* Dictionary → access by key
